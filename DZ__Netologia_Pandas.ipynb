Exercise 1

Determine which movie received the most 5.0 ratings

```python
#Задание №1
from xlrd.biffh import unicode
df = pd.read_table('ratings.dat', names = ['UserID', 'MovieIDs', 'Ratings', 'Timestamp'], sep = '::')
df.head()
```

```python
data = df[['MovieIDs', 'Ratings']]
data = data[ data.Ratings == 5 ]
best_movie = data['MovieIDs'].value_counts()
best_movie
# Фильм с MovieIDs = 2858 получил больше всех оценок 5, а именно 1963.
# The movie with MovieIDs = 2858 received the most 5 ratings, namely 1963
```

Посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 год. Не учитывайте в расчётах отрицательные значения quantity.

Exercise 2

Calculate the total consumption of the Baltic countries (Latvia, Lithuania and Estonia) of categories 4, 12 and 21 for the period from 2005 to 2010. Do not take negative quantity values into account in your calculations.



```python
#Задание №2
df = pd.read_csv('power.csv')
df.head()
```

```python
data_ = df[(df.quantity > 0) & ((df.category == 4) | (df.category == 12) | (df.category == 21)) & ((df.year >=2005) & (df.year <= 2010))]
data_c = data_[(data_['country'] == 'Latvia') | (data_['country'] == 'Lithuania') | (data_['country'] == 'Estonia') ]
data_c.quantity.sum()
# Сумарное потребление стран Прибалтики с 2005 по 2010 год = 240580.0
# Total consumption of the Baltic countries from 2005 to 2010 = 240580.0
```
